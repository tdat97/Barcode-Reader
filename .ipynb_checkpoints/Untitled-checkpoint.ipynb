```python
import pyzbar.pyzbar as pyzbar  # pip install pyzbar
import numpy as np              # pip install numpy
import cv2                      # pip install opencv-python

# 바코드 탐지하는 엔진 (바코드 및 QR코드 탐지)
def decode(im):
    # Find barcodes and QR codes
    decodedObjects = pyzbar.decode(im)

    # Print results
    for obj in decodedObjects:
        print('Type : ', obj.type)
        print('Data : ', obj.data, '\n')

    return decodedObjects
```

```python
def im_rotate(img, degree):
    h, w = img.shape[:-1]
    centerRotatePT = int(w / 2), int(h / 2)
    new_h, new_w = h, w
    rotatefigure = cv2.getRotationMatrix2D(centerRotatePT, degree, 1)
    result = cv2.warpAffine(img, rotatefigure, (new_w, new_h), )
    return result
```

```python
def apply_clahe(img): ###########
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2Lab)
    clahe = cv2.createCLAHE(clipLimit=0.5, tileGridSize=(8,8))
    lab[..., 0] = clahe.apply(lab[..., 0])
    img = cv2.cvtColor(lab, cv2.COLOR_Lab2BGR)
    return img
```

```python
def display(im, decodedObjects):
    # Loop over all decoded objects
    for decodedObject in decodedObjects:
        points = decodedObject.polygon
        print(points)

        # If the points do not form a quad, find convex hull
        if len(points) > 4:
            hull = cv2.convexHull(np.array([point for point in points], dtype=np.float32))
            hull = list(map(tuple, np.squeeze(hull)))
        else:
            hull = points;

        # Number of points in the convex hull
        n = len(hull)
        hull = np.array(hull, dtype=np.int32)

        # Draw the convext hull
        cv2.polylines(im, [hull], True, (255, 0, 0), 3)
        # for j in range(0, n):
        #     cv2.line(im, hull[j], hull[(j + 1) % n], (255, 0, 0), 3)

    # Display results
    cv2.imshow("Results", im);
    cv2.waitKey(0);
    cv2.destroyAllWindows()

# 파일명 zbar.jpg의 이미지에서 바코드를 탐지하면 해당 코드를 리턴
# Mainq
if __name__ == '__main__':
    # path = "C:/_MAIN/dataset/barcode/bar00.jpg"
    # path = "barcode.png"
    path = "barbar.png"
    im = cv2.imread(path)
    # im = cv2.resize(im, dsize=(0,0), fx=0.4, fy=0.4)
    im0 = im.copy()
    im = cv2.copyMakeBorder(im, 300, 300, 300, 300, cv2.BORDER_CONSTANT, value=0)
    # im = im_rotate(im, 90)
    # im = apply_clahe(im)
    # im = cv2.GaussianBlur(im, (0,0), 1)
    # sharpen_kernel = np.array([[0,-1,0], 
    #                            [-1,5,-1], 
    #                            [0,-1,0]])
    # im = cv2.filter2D(im, -1, sharpen_kernel2)

    decodedObjects = decode(im)
    display(im, decodedObjects)
```

```python
degs = []

for line in lines:
    xyxy = np.array(line)[0]
    dx, dy = xyxy[2:] - xyxy[:2]
    
    if dx == 0: degs.append(90); continue
    
    deg = np.arctan(dy/dx) / np.pi * 180
    if deg < 0: deg += 180
    degs.append(deg)
    
sorted(degs)
```
